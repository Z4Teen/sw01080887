```python
import requests
from bs4 import BeautifulSoup
import csv
from datetime import datetime
import re
import json
```

```python
#url to get the review from. in this case its iphone 14 from shopee
url = 'https://shopee.com.my/Apple-iPhone-14-i.304504082.19244660196?sp_atk=b718c5b0-1b8a-46a2-8b6f-99e75c14424e&xptdk=b718c5b0-1b8a-46a2-8b6f-99e75c14424e'

r = re.search(r'i\.(\d+)\.(\d+)', url)
if r:
    shop_id, item_id = r.group(1), r.group(2)
else:
    print("Invalid URL")
    exit()

offset = 0
while True:
    #using the shopee api here
    ratings_url = f'https://shopee.com.my/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0'
    #assign it in data variable
    data = requests.get(ratings_url).json()

    #loop for display to see if we're getting the correct data
    for rating in data['data']['ratings']:
        print(rating['author_username'])
        print(rating['comment'])
        ctime_timestamp = rating['ctime'] #ctime here is in json format and need to be parsed into proper date
        ctime_date = datetime.fromtimestamp(ctime_timestamp).strftime('%Y-%m-%d %H:%M:%S')
        print(ctime_date)
        print('-' * 80)


    if len(data['data']['ratings']) < 20:
        break

    offset += 20
```

```python
import re
import json
import requests
import csv


r = re.search(r'i\.(\d+)\.(\d+)', url)
if r:
    shop_id, item_id = r.group(1), r.group(2)
else:
    print("Invalid URL")
    exit()

offset = 0
csv_file = 'shopee_reviews.csv'  # CSV file to store the comments
max_pages = 5  # Limit to fetch reviews from 5 pages
csv_header = ['Reviewer', 'Comment', 'Date']


with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(csv_header)
    page_count = 0  # Counter to track the number of pages processed


    while page_count < max_pages:
        ratings_url = f'https://shopee.com.my/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0'
        data = requests.get(ratings_url).json()

        ratings_data = data.get('data', {}).get('ratings')  # get the ratings data

        if not ratings_data:
            print("No more reviews to fetch.")
            break

        #loop to assign variable to be stored in csv
        for rating in ratings_data:
            reviewer = rating['author_username']
            comment = rating['comment']
            ctime_timestamp = rating['ctime']
            ctime_date = datetime.fromtimestamp(ctime_timestamp).strftime('%Y-%m-%d %H:%M:%S')

            # Write to CSV
            writer.writerow([reviewer, comment, ctime_date])

        offset += len(ratings_data)  # Move to the next set of reviews
        page_count += 1  # Increment the page count

print(f"Comments saved to {csv_file}")
```
